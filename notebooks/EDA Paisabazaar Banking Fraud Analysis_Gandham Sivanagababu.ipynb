### Know Your Data EDA Class

````python
# ===============================================
# CLASS 1: Know Your Data
# ===============================================

class KnowYourData:
    """
    Performs initial dataset understanding and structural sanity checks.

    This class is intentionally READ-ONLY and exploratory in nature.
    It is designed to answer the question:
        "Is this dataset structurally sound and ready for analysis?"

    Design principles:
    - No mutation of the input DataFrame
    - No feature engineering or transformations
    - No dtype enforcement or imputation
    - Pure inspection and diagnostics only

    This separation ensures that data understanding remains
    independent from data wrangling and modeling logic.
    """

    def __init__(self, df: pd.DataFrame):
        """
        Initialize with the raw input DataFrame.

        Parameters:
        df : pd.DataFrame
            The raw dataset to be inspected. This DataFrame
            is never modified within this class.
        """
        self.df = df

    # --------------------------------------------------
    # Internal helpers (NOT part of public API)
    # --------------------------------------------------

    def _numeric_cols(self):
        """
        Identify numeric columns for statistical inspection.

        Returns:
        list
            Column names inferred as numeric.
        """
        return self.df.select_dtypes(include=[np.number]).columns.tolist()

    def _categorical_cols(self):
        """
        Identify categorical columns based on dtype.

        Returns:
        list
            Column names inferred as categorical or object types.
        """
        return self.df.select_dtypes(include=["object", "category"]).columns.tolist()

    def _unique_counts(self):
        """
        Compute unique value counts per column.

        Returns:
        pd.Series
            Series containing unique value counts for each column.
        """
        return self.df.nunique(dropna=False)

    # --------------------------
    # Method 1: First structural view
    # --------------------------

    def first_view(self):
        """
        Provide an initial, high-level snapshot of the dataset.

        This method surfaces:
        - Sample records (head and tail)
        - Dataset shape and column listing
        - Structural red flags such as duplicate or constant columns
        - Memory footprint
        - Basic numeric summaries
        - Sample categorical distributions for low-cardinality features
        """

        print("==== Dataset Preview (First 5 Rows) ====")
        display(self.df.head())

        print("\n==== Dataset Tail (Last 3 Rows) ====")
        display(self.df.tail(3))

        print("\n==== Dataset Shape ====")
        print(f"Rows: {self.df.shape[0]} \t Columns: {self.df.shape[1]}")

        print("\n==== Columns ====")
        print(list(self.df.columns))

        # Detect duplicate column names, which can break downstream logic
        print("\n==== Duplicate Column Names Check ====")
        dup_cols = self.df.columns[self.df.columns.duplicated()].tolist()
        print("Duplicate columns:" if dup_cols else "None found", dup_cols)

        # Detect columns with no variance (often data issues or placeholders)
        print("\n==== Constant (Single-Valued) Columns Check ====")
        const_cols = [c for c in self.df.columns if self.df[c].nunique(dropna=False) <= 1]
        print("Constant columns:" if const_cols else "No constant columns found", const_cols)

        # Memory footprint helps assess feasibility for in-memory analysis
        print("\n==== Memory Usage (MB) ====")
        mem_mb = self.df.memory_usage(deep=True).sum() / (1024 ** 2)
        print(f"{mem_mb:.2f} MB")

        # Numeric summary to understand scale, spread, and potential anomalies
        print("\n==== Quick Numeric Summary ====")
        num_cols = self._numeric_cols()
        if num_cols:
            display(self.df[num_cols].describe().T)
        else:
            print("No numeric columns found.")

        # Display representative categorical distributions (limited to avoid noise)
        print("\n==== Sample Categorical Value Counts (Low Cardinality) ====")
        shown = 0
        for c in self._categorical_cols():
            try:
                uniq = self.df[c].nunique(dropna=False)
                if uniq <= 30:
                    print(f"\n-- {c} (unique: {uniq}) --")
                    display(self.df[c].value_counts(dropna=False).head(10))
                    shown += 1
                if shown >= 6:
                    break
            except Exception:
                continue

        if shown == 0:
            print("No low-cardinality categorical columns found.")

    # --------------------------
    # Method 2: Dataset shape & uniqueness overview
    # --------------------------

    def rows_columns_count(self):
        """
        Summarize dataset size and infer semantic properties of columns.

        This method helps:
        - Confirm dataset dimensionality
        - Identify categorical candidates via low cardinality
        - Detect potential identifier columns via high uniqueness
        """

        print("==== Dataset Shape ====")
        print(f"Rows: {self.df.shape[0]} \t Columns: {self.df.shape[1]}")

        unique_counts = self._unique_counts()

        print("\n==== Unique Value Counts (Columns <= 50 unique) ====")
        for col, cnt in unique_counts.sort_values().items():
            if cnt <= 50:
                print(f"{col}: {cnt} unique values")

        # Heuristic: low-cardinality columns are strong categorical candidates
        categorical_candidates = [c for c, cnt in unique_counts.items() if cnt <= 30]

        print("\n==== Auto-Detected Categorical Candidates (<=30 unique) ====")
        print(categorical_candidates)

        # Heuristic: columns with near-unique values often represent identifiers
        possible_id_cols = [
            c for c, cnt in unique_counts.items()
            if cnt > 0.9 * len(self.df)
        ]

        print("\n==== Possible ID Columns (90%+ unique) ====")
        print(possible_id_cols)

    # --------------------------
    # Method 3: Dataset information
    # --------------------------

    def dataset_info(self):
        """
        Display structural metadata and comprehensive summary statistics.

        Includes:
        - Data types and non-null counts
        - Combined numeric and categorical descriptive statistics
        """

        print("\n==== Dataset Info ====")
        self.df.info()

        print("\n==== Full Statistical Summary (Numeric + Categorical) ====")
        display(self.df.describe(include="all"))

    # --------------------------
    # Method 4: Duplicate rows check
    # --------------------------

    def duplicate_rows(self):
        """
        Identify duplicate rows without altering the dataset.

        Duplicate handling decisions are deferred to data wrangling,
        where business context can be applied.
        """
        duplicate_count = self.df.duplicated().sum()
        print(f"\n==== Duplicate Rows Found: {duplicate_count} ====")

    # --------------------------
    # Method 5: Missing values
    # --------------------------

    def missing_values(self):
        """
        Report missing values per column.

        This method does not impute or modify data.
        It only surfaces completeness information.
        """

        missing = self.df.isnull().sum()

        print("\n==== Missing Values Per Column ====")
        display(missing)

        missing_filtered = missing[missing > 0].sort_values(ascending=False)

        print("\n==== Columns with Missing Values (Filtered) ====")
        if missing_filtered.empty:
            print("No missing values detected.")
        else:
            display(missing_filtered)

    # --------------------------
    # Method 6: Missing value visualization
    # --------------------------

    def visualize_missing(self):
        """
        Visualize missing value patterns when present.

        Visualization is explicitly skipped when no missing
        values exist, avoiding unnecessary plots.
        """

        missing = self.df.isnull().sum()
        missing_filtered = missing[missing > 0]

        if missing_filtered.empty:
            print("No missing values to visualize (explicitly skipped).")
            return

        plt.figure(figsize=(10, 4))
        missing_filtered.sort_values(ascending=False).plot(kind="bar")
        plt.title("Missing Values Count per Column")
        plt.ylabel("Missing Count")
        plt.xlabel("Columns")
        plt.xticks(rotation=45)
        plt.show()

        print("\nMissing Data Visualization — Bar")
        msno.bar(self.df)
        plt.show()

        print("\nMissing Data Visualization — Matrix")
        msno.matrix(self.df)
        plt.show()
````

### Understand your Vaiables EDA Class

````python
# ===============================================
# CLASS 2: Understand Your Variables
# ===============================================

class UnderstandVariables:
    """
    Performs variable-level inspection and diagnostic analysis.

    This class focuses on understanding individual features in terms of:
    - data types and cardinality
    - distributional characteristics
    - uniqueness patterns

    Important scope boundaries:
    - No feature engineering
    - No transformations or cleaning
    - No encoding or normalization
    - Diagnostics only, not modeling decisions

    Distributional statistics (skewness, kurtosis) are reported
    to inform interpretation of later analyses, not to justify
    transformations at this stage.
    """

    def __init__(self, df: pd.DataFrame, protected_ids: list | None = None):
        """
        Initialize with the dataset for variable-level inspection.

        Parameters:
        df : pd.DataFrame
            Input dataset to be inspected. The DataFrame is treated
            as read-only within this class.
        protected_ids : list, optional
            Column names that represent identifiers and should be
            excluded from numeric and categorical diagnostics.
        """
        # Intentionally not copying to avoid unnecessary memory overhead
        # and to signal read-only intent.
        self.df = df
        self.protected_ids = protected_ids or []

    # --------------------------------------------------
    # Internal helpers (schema-on-read)
    # --------------------------------------------------

    def _numeric_cols(self):
        """
        Identify numeric columns excluding protected identifier fields.
        """
        return [
            c for c in self.df.select_dtypes(include=[np.number]).columns
            if c not in self.protected_ids
        ]

    def _categorical_cols(self):
        """
        Identify categorical columns excluding protected identifier fields.
        """
        return [
            c for c in self.df.select_dtypes(include=["object", "category"]).columns
            if c not in self.protected_ids
        ]

    # --------------------------
    # Method 1: Column overview
    # --------------------------

    def columns_overview(self):
        """
        Provide a structural overview of dataset columns.

        Surfaces:
        - total column count
        - column names
        - data types
        - missing value percentages
        - memory footprint per column
        """

        print("==== Total Columns ====")
        print(len(self.df.columns))

        print("\n==== List of Columns ====")
        for col in self.df.columns:
            print(f"- {col}")

        display(pd.DataFrame({"Column Name": self.df.columns}))

        summary = pd.DataFrame({
            "dtype": self.df.dtypes,
            "missing %": self.df.isnull().mean() * 100,
            "memory (KB)": self.df.memory_usage(deep=True) / 1024
        })

        print("\n==== Column Summary (dtype, missing %, memory KB) ====")
        display(summary)

        return self.df

    # --------------------------
    # Method 2: Distribution diagnostics
    # --------------------------

    def dataset_describe(self):
        """
        Display descriptive statistics and distribution diagnostics.

        Includes:
        - numeric summary statistics
        - skewness to understand asymmetry
        - kurtosis to understand tail behavior
        - categorical frequency summaries

        These diagnostics support interpretation of EDA findings
        and do not imply data issues by themselves.
        """

        numeric_cols = self._numeric_cols()
        categorical_cols = self._categorical_cols()

        if numeric_cols:
            print("\n==== Numeric Summary ====")
            display(self.df[numeric_cols].describe().T)

            print("\n==== Skewness of Numeric Columns ====")
            display(self.df[numeric_cols].skew())

            print("\n==== Kurtosis of Numeric Columns ====")
            display(self.df[numeric_cols].kurtosis())
        else:
            print("No numeric columns found.")

        if categorical_cols:
            print("\n==== Categorical Summary ====")
            display(self.df[categorical_cols].describe())
        else:
            print("No categorical columns found.")

        return self.df

    # --------------------------
    # Method 3: Cardinality and uniqueness analysis
    # --------------------------

    def unique_values_analysis(self, preview_limit: int = 10):
        """
        Analyze feature cardinality and unique value patterns.

        This method helps:
        - identify high-cardinality features
        - detect low-cardinality categorical candidates
        - surface potential encoding or grouping considerations

        Parameters:
        preview_limit : int
            Number of unique values to preview for high-cardinality features.
        """

        print("==== Unique Value Count Per Column ====")
        unique_counts = self.df.nunique(dropna=False).sort_values()
        display(unique_counts)

        print("\n==== Sample Unique Values Preview ====")

        for col in self.df.columns:
            nunique = self.df[col].nunique(dropna=False)
            print(f"\n{col} ({self.df[col].dtype}) - {nunique} unique values")

            if nunique <= 15:
                print("Full unique values:")
                display(self.df[col].unique())
            else:
                print(f"First {preview_limit} unique values:")
                display(self.df[col].unique()[:preview_limit])

        print("\n==== High-Cardinality Columns (>1000 unique) ====")
        display(unique_counts[unique_counts > 1000])

        print("\n==== Low-Cardinality Columns (<20 unique) ====")
        display(unique_counts[unique_counts < 20])

        return self.df
````

````python
# ================================================================
# CLASS 3: Data Wrangling / Cleaning
# ================================================================

class DataWrangler:
    """
    Performs deterministic, non-destructive data cleaning operations.

    This class represents a transformation boundary in the pipeline and
    therefore owns its own copy of the dataset.

    Responsibilities:
    - Remove duplicate rows
    - Apply light, non-semantic text normalization
    - Safely coerce known numeric-by-definition columns
    - Detect zero-variance numeric features (reporting only)
    - Detect outliers using IQR (analysis only, no removal)

    Explicitly excluded:
    - Identifier handling (delegated to IDColumnsManager)
    - Feature engineering
    - Encoding, scaling, or modeling logic
    """

    def __init__(self, df: pd.DataFrame):
        """
        Initialize the DataWrangler with a defensive copy.

        Parameters:
        df : pd.DataFrame
            Input dataset. A copy is created to prevent
            mutation of upstream pipeline state.
        """
        self.df = df.copy()
        self.outlier_counts = None
        self.zero_variance_cols = None

        print("\n================= DATA WRANGLING INITIALIZED =================")
        print(f"Input shape: {self.df.shape}")

    # ------------------------------------------------------------
    # Step 1: Remove duplicate rows
    # ------------------------------------------------------------

    def remove_duplicates(self):
        """
        Remove exact duplicate rows.

        Duplicate handling is performed early to avoid
        propagating redundant records through the pipeline.
        """
        print("\n---- Step 1: Removing Duplicates ----")

        duplicate_count = self.df.duplicated().sum()
        print(f"Duplicate rows detected: {duplicate_count}")

        if duplicate_count > 0:
            self.df.drop_duplicates(inplace=True)
            print("Duplicate rows removed.")
        else:
            print("No duplicate rows found.")

        print(f"Shape after de-duplication: {self.df.shape}")
        return self

    # ------------------------------------------------------------
    # Step 2: Clean text columns (light normalization)
    # ------------------------------------------------------------

    def clean_text_columns(self):
        """
        Apply light, non-semantic normalization to text columns.

        Operations:
        - Strip leading/trailing whitespace
        - Collapse multiple spaces
        - Normalize case

        This step improves consistency for grouping and comparison
        without altering semantic meaning.
        """
        print("\n---- Step 2: Cleaning Text Columns ----")

        text_columns = self.df.select_dtypes(include=["object"]).columns.tolist()
        print(f"Text columns detected: {text_columns}")

        for col in text_columns:
            self.df[col] = (
                self.df[col]
                .astype(str)
                .str.strip()
                .str.replace(r"\s+", "_", regex=True)
                .str.lower()
            )
            print(f"Normalized text column: {col}")

        print("Text normalization completed.")
        return self

    # ------------------------------------------------------------
    # Step 3: Convert known numeric-by-definition columns
    # ------------------------------------------------------------

    def convert_numeric_columns(self):
        """
        Safely coerce business-defined numeric columns.

        These columns are expected to be numeric by domain definition,
        even if ingested as strings due to upstream data issues.
        """
        print("\n---- Step 3: Converting Known Numeric Columns ----")

        numeric_columns = [
            "Outstanding_Debt",
            "Interest_Rate",
            "Credit_Utilization_Ratio",
            "Num_of_Loan",
            "Age",
            "Annual_Income",
            "Monthly_Inhand_Salary"
        ]

        for col in numeric_columns:
            if col in self.df.columns:
                before_nulls = self.df[col].isna().sum()
                self.df[col] = pd.to_numeric(self.df[col], errors="coerce")
                after_nulls = self.df[col].isna().sum()

                print(
                    f"Converted '{col}' to numeric | "
                    f"NaNs before: {before_nulls}, after: {after_nulls}"
                )
            else:
                print(f"Skipped (column not found): {col}")

        return self

    # ------------------------------------------------------------
    # Step 3B: Optional numeric coercion hook
    # ------------------------------------------------------------

    def coerce_numeric_columns(self, cols: list):
        """
        Coerce additional columns to numeric on demand.

        This hook supports extensibility without modifying
        the core wrangling logic.
        """
        print("\n---- Step 3B: Coercing Custom Numeric Columns ----")

        for col in cols:
            if col in self.df.columns:
                self.df[col] = pd.to_numeric(self.df[col], errors="coerce")
                print(f"Coerced to numeric: {col}")
            else:
                print(f"Skipped (column not found): {col}")

        return self

    # ------------------------------------------------------------
    # Step 4: Detect zero-variance numeric columns
    # ------------------------------------------------------------

    def detect_zero_variance(self):
        """
        Detect numeric columns with zero variance.

        Zero-variance features provide no analytical value and
        are reported for downstream exclusion rather than mutated.
        """
        print("\n---- Step 4: Detecting Zero-Variance Columns ----")

        numeric_cols = self.df.select_dtypes(include=[np.number]).columns
        self.zero_variance_cols = [c for c in numeric_cols if self.df[c].std() == 0]

        if self.zero_variance_cols:
            print(f"Zero-variance numeric columns detected: {self.zero_variance_cols}")
        else:
            print("No zero-variance numeric columns detected.")

        return self

    # ------------------------------------------------------------
    # Step 5: Detect outliers using IQR (reporting only)
    # ------------------------------------------------------------

    def detect_iqr_outliers(self):
        """
        Detect potential outliers using the IQR method.

        This step is strictly diagnostic:
        - No rows are removed
        - No values are capped or modified
        """
        print("\n---- Step 5: IQR-Based Outlier Detection ----")

        numeric_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
        outlier_counts = {}

        for col in numeric_cols:
            s = self.df[col].dropna()
            if s.empty:
                continue

            Q1 = s.quantile(0.25)
            Q3 = s.quantile(0.75)
            IQR = Q3 - Q1

            if IQR == 0:
                outlier_counts[col] = 0
                continue

            lower = Q1 - 1.5 * IQR
            upper = Q3 + 1.5 * IQR

            outlier_counts[col] = int(((s < lower) | (s > upper)).sum())

        self.outlier_counts = outlier_counts

        print("\nTop 10 Columns with Maximum Outliers:")
        display(
            pd.DataFrame(
                sorted(outlier_counts.items(), key=lambda x: x[1], reverse=True)[:10],
                columns=["Feature", "Outlier_Count"]
            )
        )

        print("Outlier detection completed (analysis only).")
        return self

    # ------------------------------------------------------------
    # Step 6: Final output
    # ------------------------------------------------------------

    def get_df(self) -> pd.DataFrame:
        """
        Return the wrangled dataset and summarize applied steps.
        """
        print("\n================= DATA WRANGLING SUMMARY =================")
        print(f"Final dataset shape: {self.df.shape}")
        print("- Duplicates handled")
        print("- Text normalized")
        print("- Numeric coercion applied")
        print("- Zero-variance features detected")
        print("- Outliers analyzed (no rows removed)")

        display(self.df.head())
        return self.df
````

###ID columns Locking

````python
# ================================================================
# Class 4: ID Column Manager
# ================================================================
class IDColumnsManager:
    """
    ID handling:
    1) Detect strict ROW IDs (primary keys)
    2) Detect ENTITY IDs (Customer_ID, SSN, Name, etc.)
    3) Combine + lock for downstream FE / EDA safety
    """

    ROW_ID_KEYWORDS = ["id", "uuid", "guid"]
    ENTITY_ID_KEYWORDS = ["customer", "ssn", "client", "name", "user"]

    def __init__(self, df: pd.DataFrame):
        self.df = df
        self.row_ids = []
        self.entity_ids = []
        self.protected_ids = []

    # --------------------------------------------------
    # STEP 1: Row-level ID detection (PRIMARY KEYS)
    # --------------------------------------------------
    def detect_row_ids(self) -> list:
        ids = []
        n = len(self.df)

        for col in self.df.columns:
            col_lower = col.lower()

            if not any(k in col_lower for k in self.ROW_ID_KEYWORDS):
                continue

            # must be nearly unique
            if self.df[col].nunique(dropna=False) / n < 0.98:
                continue

            # no missing values
            if self.df[col].isnull().any():
                continue

            # avoid floats pretending to be IDs
            if pd.api.types.is_float_dtype(self.df[col]):
                continue

            ids.append(col)

        self.row_ids = ids
        return ids

    # --------------------------------------------------
    # STEP 2: Entity ID detection (PROTECTED IDENTIFIERS)
    # --------------------------------------------------
    def detect_entity_ids(self) -> list:
        ids = []

        for col in self.df.columns:
            col_lower = col.lower()

            if any(k in col_lower for k in self.ENTITY_ID_KEYWORDS):
                # additional guard: entity IDs usually have high cardinality
                nunique_ratio = self.df[col].nunique(dropna=False) / len(self.df)

                if nunique_ratio > 0.1:  # avoids low-cardinality categorical misuse
                    ids.append(col)

        self.entity_ids = ids
        return ids

    # --------------------------------------------------
    # STEP 3: Combine
    # --------------------------------------------------
    def detect(self) -> list:
        self.detect_row_ids()
        self.detect_entity_ids()

        self.protected_ids = sorted(set(self.row_ids + self.entity_ids))
        return self.protected_ids

    # --------------------------------------------------
    # STEP 4: Lock (explicit side effect)
    # --------------------------------------------------
    def lock(self) -> pd.DataFrame:
        print("\n---- ID COLUMN LOCKING ----")
        for col in self.protected_ids:
            self.df[col] = self.df[col].astype("object")
            print(f"Locked ID column: {col}")
        return self.df

    # --------------------------------------------------
    # Utilities for FE / EDA
    # --------------------------------------------------
    def exclude_from_numeric(self, numeric_cols: list) -> list:
        return [c for c in numeric_cols if c not in self.protected_ids]

    def exclude_from_categorical(self, categorical_cols: list) -> list:
        return [c for c in categorical_cols if c not in self.protected_ids]

    def get_ids(self) -> list:
        return self.protected_ids

# ----------------------------------
# Step 1: Detect & Lock ID Columns
# ----------------------------------
id_manager = IDColumnsManager(df_clean)

# Auto-detect ID columns
ID_COLUMNS = id_manager.detect() #protected: Row Identifier + Entity Identifier
print("Protected ID Columns:", ID_COLUMNS)

# Lock ID columns (convert to object type)
df_clean = id_manager.lock()

````

Following data wrangling, identifier columns were explicitly detected and locked using a dedicated ID management component. Row-level identifiers and entity-level identifiers such as customer ID, SSN, and name were automatically identified using a combination of naming heuristics and cardinality checks.

These identifier columns were retained strictly for traceability and auditing purposes and were explicitly excluded from numeric and categorical analysis to prevent leakage. Locking identifiers ensured that all downstream exploratory and modeling insights were derived solely from behavioral and financial attributes.

### Feature Engineering

````python
# ================================================================
# Class 5: Feature Engineering
# ================================================================
class FeatureEngineering:
    """
    Deterministic, reusable feature transformations applied prior to EDA
    and downstream modeling.

    Design principles:
    - Defensive copy: input dataframe is never mutated
    - ID-safe: identifier columns are never transformed or derived from
    - Deterministic: no randomness, fully reproducible outputs
    - Chainable API: transformations can be composed step-by-step
    """

    def __init__(self, df: pd.DataFrame, id_columns: list = None):
        """
        Initialize feature engineering pipeline.

        Parameters:
        - df: cleaned dataframe
        - id_columns: list of identifier columns to be excluded from feature logic
        """
        self.df = df.copy()
        self.id_columns = id_columns if id_columns else []

    # ---------------------------------------------------------
    # Helper: Exclude ID columns defensively
    # ---------------------------------------------------------
    def _exclude_ids_from_numeric(self, cols: list) -> list:
        """
        Filter numeric columns to ensure ID columns are excluded.
        """
        return [c for c in cols if c not in self.id_columns]

    def _exclude_ids_from_categorical(self, cols: list) -> list:
        """
        Filter categorical columns to ensure ID columns are excluded.
        """
        return [c for c in cols if c not in self.id_columns]

    # ---------------------------------------------------------
    def map_credit_score_to_numeric(
        self, col: str = "Credit_Score", out_col: str = "Credit_Score_Num"
    ) -> "FeatureEngineering":
        """
        Map credit score categories to an ordered numeric scale.

        Ordinal semantics preserved:
        Poor < Standard < Good

        Behavior:
        - Uses explicit domain mapping
        - Unknown or malformed values default to the median risk category
        """
        if col not in self.df.columns:
            raise KeyError(f"Column '{col}' not found in DataFrame")

        # Avoid recomputation if already present and valid
        if out_col in self.df.columns and pd.api.types.is_numeric_dtype(self.df[out_col]):
            return self

        mapping = {
            "poor": 1,
            "standard": 2,
            "good": 3
        }

        s = self.df[col].astype(str).str.lower()
        self.df[out_col] = s.map(mapping)

        # Explicit fallback for unexpected values
        unknown_mask = self.df[out_col].isna() & s.notna()
        if unknown_mask.any():
            # assign median score instead of random encoding
            self.df.loc[unknown_mask, out_col] = 2

        self.df[out_col] = pd.to_numeric(self.df[out_col], errors="coerce")
        return self

    # ---------------------------------------------------------
    def convert_credit_history_age_to_years(
        self, src_col: str = "Credit_History_Age", out_col: str = "Credit_History_Years"
    ) -> "FeatureEngineering":
        """
        Normalize credit history duration into numeric years.

        Supported formats:
        - '10 Years 2 Months'
        - '120 months'
        - numeric month values

        Output:
        - Floating-point years
        """

        if src_col not in self.df.columns:
            raise KeyError(f"Column '{src_col}' not found.")

        pattern_years = re.compile(r"(\d+)\s*(?:years?|yrs?|y)", re.IGNORECASE)
        pattern_months = re.compile(r"(\d+)\s*(?:months?|mons?|m)", re.IGNORECASE)

        def _convert(val):
            if pd.isna(val):
                return np.nan
            if isinstance(val, (int, float)):
                return float(val) / 12

            s = str(val).strip().lower()
            years = pattern_years.search(s)
            months = pattern_months.search(s)

            total = 0.0
            if years:
                total += float(years.group(1))
            if months:
                total += float(months.group(1)) / 12

            return total if total > 0 else np.nan

        self.df[out_col] = self.df[src_col].apply(_convert)
        return self

    # ---------------------------------------------------------
    def compute_debt_to_income_ratio(
        self,
        emi_col="Total_EMI_per_month",
        annual_col="Annual_Income",
        monthly_col="Monthly_Inhand_Salary",
        out_col="Debt_To_Income_Ratio"
    ) -> "FeatureEngineering":
        """
        Compute Debt-to-Income (DTI) ratio as a percentage.

        Strategy:
        - Compute DTI using both annual and monthly income sources
        - Select the version with higher data completeness
        - Avoid division by zero and invalid income values
        """

        if emi_col not in self.df.columns:
            raise KeyError(f"{emi_col} not found.")

        self.df[emi_col] = pd.to_numeric(self.df[emi_col], errors="coerce")

        # Annual-based DTI
        self.df["DTI_annual_pct"] = np.where(
            self.df.get(annual_col, 0) > 0,
            (self.df[emi_col] * 12) / self.df[annual_col] * 100,
            np.nan
        )

        # Monthly-based DTI
        self.df["DTI_monthly_pct"] = np.where(
            self.df.get(monthly_col, 0) > 0,
            self.df[emi_col] / self.df[monthly_col] * 100,
            np.nan
        )

        # Choose the more complete version
        if self.df["DTI_annual_pct"].notna().sum() >= self.df["DTI_monthly_pct"].notna().sum():
            self.df[out_col] = self.df["DTI_annual_pct"]
        else:
            self.df[out_col] = self.df["DTI_monthly_pct"]

        return self

    # ---------------------------------------------------------
    def create_default_flag(
        self, delayed_col="Num_of_Delayed_Payment", out_col="Has_Default"
    ) -> "FeatureEngineering":
        """
        Create a binary default indicator.

        Definition:
        - 1: at least one delayed payment observed
        - 0: no delayed payments
        """

        if delayed_col not in self.df.columns:
            raise KeyError(f"{delayed_col} not found.")

        self.df[out_col] = np.where(
            pd.to_numeric(self.df[delayed_col], errors="coerce") > 0, 1, 0
        )

        return self
    # ---------------------------------------------------------
    def get_df(self) -> pd.DataFrame:
        """Return the current DataFrame after all transformations."""
        return self.df
````

A key transformation involved converting the credit history age variable from its original non-numeric representation into a numeric tenure-based measure expressed in years. This transformation enables meaningful comparison of credit maturity across customers and supports analysis of stability and long-term credit behavior.

Additional engineered features such as an ordinal credit score mapping, debt-to-income ratio, and default indicator were created using deterministic, domain-driven logic. All feature transformations were designed to preserve semantic meaning, avoid identifier leakage, and maintain reproducibility.

### Schema Evolution

````python
# ============================================================
# Helper: Extract schema info
# ============================================================
def _extract_schema(df: pd.DataFrame) -> pd.DataFrame:
    """
    Extract column-level schema metadata for a dataframe.

    Purpose:
    - Capture structural properties of a dataframe snapshot
    - Enable before/after comparison during data pipeline stages
    - Support schema governance and auditability

    Collected attributes:
    - Column name
    - Data type
    - Non-null count
    - Null count
    - Unique value count (including NaN)

    Notes:
    - This function is read-only
    - Used internally by schema_diff()
    """
    return pd.DataFrame({
        "column_name": df.columns,
        "dtype": df.dtypes.astype(str),
        "non_null_count": df.notna().sum().values,
        "null_count": df.isna().sum().values,
        "unique_count": df.nunique(dropna=False).values
    })


def schema_diff(df_before, df_after, before_name, after_name):
    """
    Compare schema differences between two dataframe states.

    Purpose:
    - Track structural drift across pipeline stages
    - Detect unintended mutations early
    - Provide transparent schema change logs

    Comparisons performed:
    - Added columns
    - Removed columns
    - Modified columns (dtype, null count, unique count)

    Parameters
    ----------
    df_before : pd.DataFrame
        Dataframe snapshot before transformation
    df_after : pd.DataFrame
        Dataframe snapshot after transformation
    before_name : str
        Human-readable label for the before state
    after_name : str
        Human-readable label for the after state

    Returns
    -------
    dict
        Structured schema diff containing:
        - source and target names
        - added columns
        - removed columns
        - modified columns dataframe
    """
    schema_before = _extract_schema(df_before)
    schema_after = _extract_schema(df_after)

    cols_before = set(schema_before["column_name"])
    cols_after = set(schema_after["column_name"])

    added_cols = sorted(cols_after - cols_before)
    removed_cols = sorted(cols_before - cols_after)
    common_cols = sorted(cols_before & cols_after)

    changes = []
    for col in common_cols:
        b = schema_before[schema_before["column_name"] == col].iloc[0]
        a = schema_after[schema_after["column_name"] == col].iloc[0]

        if (
            b["dtype"] != a["dtype"] or
            b["null_count"] != a["null_count"] or
            b["unique_count"] != a["unique_count"]
        ):
            changes.append({
                "column": col,
                "dtype_before": b["dtype"],
                "dtype_after": a["dtype"],
                "nulls_before": int(b["null_count"]),
                "nulls_after": int(a["null_count"]),
                "unique_before": int(b["unique_count"]),
                "unique_after": int(a["unique_count"])
            })

    return {
        "from": before_name,
        "to": after_name,
        "added_columns": added_cols,
        "removed_columns": removed_cols,
        "modified_columns": pd.DataFrame(changes)
    }


def _display_schema_diff(diff: dict):
    """
    Render schema differences in a readable, executive-friendly format.

    Purpose:
    - Make schema evolution visually inspectable
    - Support notebook-based reviews and audits
    - Avoid silent structural changes

    Notes:
    - Read-only display helper
    - Intended for notebook and reporting contexts
    """

    print(f"\nFROM: {diff['from']} → TO: {diff['to']}")

    print("\n - Added Columns:")
    print(diff["added_columns"] if diff["added_columns"] else "None")

    print("\n - Removed Columns:")
    print(diff["removed_columns"] if diff["removed_columns"] else "None")

    print("\n - Modified Columns:")
    if not diff["modified_columns"].empty:
        display(diff["modified_columns"])
    else:
        print("None")


def log_full_schema_evolution(df_raw, df_clean, df_fe):
    """
    Log full schema evolution across the data pipeline.

    Pipeline stages covered:
    - Raw ingestion to cleaned dataset
    - Cleaned dataset to feature engineered dataset

    Purpose:
    - Provide end-to-end schema lineage visibility
    - Enable deterministic debugging and structured review
    - Support data governance and audit requirements
    - Ensure all schema transformations are intentional and traceable

    Usage:
    - Invoke once after the data cleaning stage is completed
    - Invoke again after feature engineering is finalized
    """

    print("\n" + "="*70)
    print("SCHEMA EVOLUTION: RAW ➜ CLEAN")
    print("="*70)

    _display_schema_diff(
        schema_diff(df_raw, df_clean, "df_raw", "df_clean")
    )

    print("\n" + "="*70)
    print("SCHEMA EVOLUTION: CLEAN ➜ FEATURE ENGINEERED")
    print("="*70)

    _display_schema_diff(
        schema_diff(df_clean, df_fe, "df_clean", "df_fe")
    )

````

To ensure full transparency and prevent unintended structural drift, schema evolution was explicitly tracked across key pipeline stages. Schema snapshots were compared between the raw dataset, the cleaned dataset, and the feature-engineered dataset.

No changes in row count, missingness, or feature cardinality were introduced at this stage, confirming that data wrangling preserved the original information content.

During the cleaned to feature-engineered transition, new analytical features were added in a controlled and deterministic manner. It provides confidence that downstream exploratory analysis and modeling are based on a stable and governed data structure, free from silent mutations or unintended data leakage.

````python
# =============================================================
# Class 6: Feature Ranking
# =============================================================

class FeatureRankingEDA:
    """
    Feature ranking utility for EDA guidance ONLY (not modeling).

    This class is used to identify and prioritize features that show
    strong association with the target variable in order to:
    - guide Univariate, Bivariate, and Multivariate EDA
    - focus visualization and analysis on impactful variables
    - prevent ID leakage and target leakage during exploration

    Important notes:
    - Operates on a defensive copy of the input DataFrame
    - Does NOT mutate df_fe or upstream data
    - Protected identifier columns are strictly excluded
    - Outputs are diagnostic, not prescriptive for modeling
    """

    def __init__(
        self,
        df: pd.DataFrame,
        target_col: str = "Credit_Score",
        max_sample: int = 8000,
        protected_ids: list = None
    ):
        """
        Initialize the feature ranking workflow.

        Parameters
        ----------
        df : pd.DataFrame
            Feature-engineered dataset used for EDA-level ranking.
        target_col : str
            Name of the target column.
        max_sample : int
            Maximum number of rows to sample for tree-based importance
            to control computation cost.
        protected_ids : list
            Columns representing identifiers to be excluded.
        """
        # Defensive copy to ensure no mutation of upstream dataset
        self.df = df.copy()
        self.target_col = target_col
        self.max_sample = int(max_sample)
        self.protected_ids = protected_ids or []

        # Internal working copy used only within this class
        self._df = self.df.copy()

        # Containers populated progressively across steps
        self.numeric_cols = []
        self.cat_cols = []
        self.num_summary = None
        self.cat_summary = None
        self.corr_with_target = None
        self.mi_series = None
        self.X_enc = None
        self.y = None
        self.rf_imp = None
        self.rank_df = None

    # ---------------------------------------------------------
    # 1) Map target to numeric (safe & explicit)
    # ---------------------------------------------------------
    def map_target_numeric(self):
        """
        Create an internal numeric representation of the target.

        Priority order:
        1) Use existing <target>_Num column if present and numeric
        2) Apply domain-aware mapping (e.g., poor < standard < good)
        3) Fallback to label encoding if no semantic mapping applies

        The numeric target is stored internally as '__target__'
        and is never written back to the original dataset.
        """
        num_col = f"{self.target_col}_Num"

        if num_col in self._df.columns and pd.api.types.is_numeric_dtype(self._df[num_col]):
            self._df["__target__"] = self._df[num_col]
            return self

        if self.target_col not in self._df.columns:
            raise KeyError(f"Target column '{self.target_col}' not found")

        mapping = {"poor": 1, "standard": 2, "good": 3}
        raw = self._df[self.target_col].astype(str).str.lower()

        mapped = raw.map(mapping)
        if mapped.notna().sum() > 0:
            self._df["__target__"] = mapped
        else:
            le = LabelEncoder()
            self._df["__target__"] = le.fit_transform(raw)

        self._df["__target__"] = pd.to_numeric(self._df["__target__"], errors="coerce")
        return self

    # ---------------------------------------------------------
    # 2) Identify numeric & categorical predictors (ID-safe)
    # ---------------------------------------------------------
    def identify_columns(self):
        """
        Identify numeric and categorical predictor columns.

        Excludes:
        - protected identifier columns
        - target column
        - derived numeric target (__target__)
        """
        exclude = set(self.protected_ids) | {
            self.target_col,
            "__target__",
            f"{self.target_col}_Num"
        }

        self.numeric_cols = [
            c for c in self._df.select_dtypes(include=np.number).columns
            if c not in exclude
        ]

        self.cat_cols = [
            c for c in self._df.select_dtypes(include=["object", "category"]).columns
            if c not in exclude
        ]

        return self

    # ---------------------------------------------------------
    # 3) Numeric univariate summary
    # ---------------------------------------------------------
    def numeric_summary_univariate(self):
        """
        Generate univariate summary statistics for numeric features.

        Captures:
        - central tendency and dispersion
        - missingness
        - skewness and kurtosis for distributional awareness
        """
        if not self.numeric_cols:
            self.num_summary = pd.DataFrame()
            return self

        desc = self._df[self.numeric_cols].describe().T
        desc["missing_count"] = self._df[self.numeric_cols].isna().sum()
        desc["missing_pct"] = desc["missing_count"] / len(self._df) * 100
        desc["skew"] = self._df[self.numeric_cols].skew()
        desc["kurtosis"] = self._df[self.numeric_cols].kurtosis()

        self.num_summary = desc
        return self

    # ---------------------------------------------------------
    # 4) Categorical univariate summary
    # ---------------------------------------------------------
    def categorical_summary_univariate(self):
        """
        Generate univariate summary statistics for categorical features.

        Captures:
        - cardinality
        - missingness
        - most frequent value and its frequency
        """
        rows = []

        for c in self.cat_cols:
            s = self._df[c].astype(object)
            vc = s.value_counts(dropna=False)

            rows.append({
                "feature": c,
                "n_unique": int(s.nunique(dropna=False)),
                "missing_count": int(s.isna().sum()),
                "missing_pct": float(s.isna().mean() * 100),
                "top_value": vc.index[0] if len(vc) else None,
                "top_freq": int(vc.iloc[0]) if len(vc) else 0
            })

        self.cat_summary = (
            pd.DataFrame(rows).set_index("feature") if rows else pd.DataFrame()
        )
        return self

    # ---------------------------------------------------------
    # 5) Correlation with target (numeric only)
    # ---------------------------------------------------------
    def correlation_with_target(self):
        """
        Compute absolute correlation between numeric features and target.

        Used as a linear association signal for EDA prioritization.
        """
        if "__target__" not in self._df.columns:
            raise RuntimeError("Call map_target_numeric() first")

        corrs = {
            c: self._df[c].corr(self._df["__target__"])
            for c in self.numeric_cols
        }

        self.corr_with_target = (
            pd.Series(corrs).abs().sort_values(ascending=False)
        )
        return self

    # ---------------------------------------------------------
    # 6) Mutual Information (EDA-level encoding)
    # ---------------------------------------------------------
    def mutual_info_features(self):
        """
        Compute mutual information scores for numeric and categorical features.

        Categorical features are ordinal-encoded strictly for MI computation.
        This encoding is EDA-only and not reused elsewhere.
        """
        if "__target__" not in self._df.columns:
            raise RuntimeError("Call map_target_numeric() first")

        features = self.numeric_cols + self.cat_cols
        if not features:
            self.mi_series = pd.Series(dtype=float)
            return self

        X = self._df[features].copy()

        if self.cat_cols:
            oe = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
            X[self.cat_cols] = oe.fit_transform(X[self.cat_cols].astype(str))

        # Simple, consistent imputation for MI computation
        X = X.fillna(X.median(numeric_only=True)).fillna(-999)
        y = self._df["__target__"].fillna(method="ffill").astype(int)

        mi = mutual_info_classif(
            X, y,
            discrete_features=[c in self.cat_cols for c in X.columns],
            random_state=42
        )

        self.mi_series = pd.Series(mi, index=X.columns).sort_values(ascending=False)
        self.X_enc = X
        self.y = y
        return self

    # ---------------------------------------------------------
    # 7) RandomForest importance (sampled)
    # ---------------------------------------------------------
    def random_forest_importance(self):
        """
        Compute RandomForest-based feature importance on a sampled dataset.

        Sampling is used to balance signal extraction with computational cost.
        """
        if self.X_enc is None or self.y is None:
            raise RuntimeError("Call mutual_info_features() first")

        n = min(self.max_sample, len(self.X_enc))
        idx = self.X_enc.sample(n=n, random_state=42).index

        Xs = self.X_enc.loc[idx]
        ys = self.y.loc[idx]

        rf = RandomForestClassifier(
            n_estimators=100,
            max_depth=6,
            random_state=42,
            n_jobs=-1
        )
        rf.fit(Xs, ys)

        self.rf_imp = (
            pd.Series(rf.feature_importances_, index=self.X_enc.columns)
            .sort_values(ascending=False)
        )
        return self

    # ---------------------------------------------------------
    # 8) Combine ranking metrics
    # ---------------------------------------------------------
    def combine_ranking_metrics(self):
        """
        Combine correlation, mutual information, and tree importance
        into a single normalized ranking score.

        Weights are chosen for EDA prioritization, not model optimization.
        """
        corr = self.corr_with_target.reindex(self.X_enc.columns).fillna(0)
        mi = self.mi_series.reindex(self.X_enc.columns).fillna(0)
        rf = self.rf_imp.reindex(self.X_enc.columns).fillna(0)

        df = pd.DataFrame({
            "corr_abs": corr,
            "mutual_info": mi,
            "rf_importance": rf
        })

        for c in df.columns:
            df[c + "_norm"] = df[c] / df[c].max() if df[c].max() > 0 else 0

        df["combined_score"] = (
            0.4 * df["corr_abs_norm"] +
            0.35 * df["mutual_info_norm"] +
            0.25 * df["rf_importance_norm"]
        )

        self.rank_df = df.sort_values("combined_score", ascending=False)
        return self

    # ---------------------------------------------------------
    # 9) Display & save
    # ---------------------------------------------------------
    def display_top_features(self, n: int = 30):
        """
        Display top-ranked features for EDA focus.
        """
        display(
            self.rank_df[
                ["corr_abs", "mutual_info", "rf_importance", "combined_score"]
            ].head(n)
        )
        return self

    def save_outputs(
        self,
        path_numeric="univariate_numeric_summary.csv",
        path_categorical="univariate_categorical_summary.csv",
        path_ranking="feature_ranking.csv"
    ):
        """
        Persist EDA summaries and feature rankings to disk.
        """
        if self.num_summary is not None:
            self.num_summary.to_csv(path_numeric)
        if self.cat_summary is not None:
            self.cat_summary.to_csv(path_categorical)
        if self.rank_df is not None:
            self.rank_df.to_csv(path_ranking)

        print(f"Saved: {path_numeric}, {path_categorical}, {path_ranking}")
        return self

    # ---------------------------------------------------------
    def get_df(self) -> pd.DataFrame:
        """
        Return the internal EDA-only DataFrame.

        Note:
        Contains the derived '__target__' column and
        should not be reused outside this class.
        """
        return self._df
````

### Exploratory Data Analysis (EDA) – Execution Principles


This Exploratory Data Analysis (EDA) stage translates prepared data into
interpretable credit risk insights while strictly preserving data integrity.

EDA in this project is treated as a **read-only analytical layer**.
No structural changes, feature creation, or dataset mutation occurs in this phase.

1. Read-Only & Non-Destructive Analysis
- The dataset used for EDA is already cleaned, governed, and feature-engineered
- All EDA operations are visual or diagnostic only
- Any binning, ordering, or transformation is applied locally within plots
- No columns are added, removed, or modified during EDA



2. Identifier Safety & Leakage Prevention
- All row-level and entity-level identifiers are explicitly protected
- Identifier columns are excluded from numeric and categorical EDA logic
- Guardrails prevent accidental plotting or inference from ID-like fields
- Ensures insights are driven only by behavioral and financial attributes



3. Risk-Oriented Analytical Structure
EDA is executed in a progressive, risk-driven sequence:

- **Univariate EDA**  
  Establishes baseline distributions, skewness, and outlier behavior for
  individual variables.

- **Bivariate EDA**  
  Examines how risk shifts when one variable conditions another, with explicit
  focus on separation across credit quality levels.

- **Multivariate EDA**  
  Used strictly for confirmation and stability checks, not for discovery or
  narrative creation.


4. Ordinal Treatment of Credit Risk
- Credit score is treated as an ordered risk outcome
- Visualizations respect ordinal semantics rather than numeric assumptions
- Comparisons focus on distributional separation, compression, and stability
  across credit quality segments



5. Controlled Complexity & Performance
- Deterministic sampling is applied for computationally expensive plots
- Plot counts are intentionally limited to avoid analytical overfitting
- Emphasis is placed on interpretability over visual density



6. Domain-Aware Visualization Standards
- Financial variables are displayed using appropriate units and scales
- Risk thresholds (e.g., Debt-to-Income bands) are overlaid where meaningful
- Encoded categorical values are rendered into human-readable labels to support
  executive-level interpretation


7. Reusable EDA Architecture
The EDA framework is implemented using modular, reusable classes:

- **BaseEDA**  
  Centralized infrastructure for sampling, styling, axis formatting, ID
  protection, and fault-tolerant plotting.

- **UnivariateEDA**  
  Distribution-level analysis for standalone risk signals.

- **BivariateEDA**  
  Pairwise analysis to explain conditional risk behavior.

- **MultivariateEDA**  
  Interaction-level confirmation aligned with senior stakeholder communication.

This structure ensures EDA outputs are consistent, auditable, and aligned with
real-world credit risk analytics standards.


````python
# ================================================================
# Class 7: Base EDA
# ================================================================
class BaseEDA:
    """
    This class serves as the single source of truth for all shared
    EDA infrastructure concerns, including:
    - Safe and consistent sampling for heavy plots
    - Axis formatting and unit standardization
    - Plot styling and layout consistency
    - Identifier column protection
    - Fault-tolerant plotting for both Matplotlib and Plotly

    All downstream EDA classes (Univariate, Bivariate, Multivariate)
    should inherit from or rely on this base to ensure consistency
    and reproducibility across visualizations.

    - Single source for:
    - Sampling
    - Axis formatting
    - Titles & units
    - Styling
    - ID protection
    - Safe plotting
    """
    # --------------------------------------------------
    # Global risk color encoding
    # --------------------------------------------------
    RISK_PALETTE = {
        "poor": "#C44E52",       # High risk
        "standard": "#DD8452",   # Medium risk
        "good": "#4C72B0"        # Low risk
    }

    # Numeric distribution colors (non-risk variables)
    NUMERIC_COLOR = "#4C72B0"        # neutral blue
    NUMERIC_KDE_COLOR = "#2F5A8A"    # darker blue for emphasis

    def __init__(
        self,
        df: pd.DataFrame,
        outputs_dir: str = "outputs",
        protected_ids: Optional[List[str]] = None,
        sample_max_rows: int = 8000
    ):
        """
        Initialize the BaseEDA context.

        Parameters
        ----------
        df : pd.DataFrame
            Input dataset for EDA.
        outputs_dir : str
            Directory where generated plots will be saved.
        protected_ids : list, optional
            Identifier columns that must never be plotted.
        sample_max_rows : int
            Maximum number of rows to sample for computationally
            expensive visualizations.
        """
        self.df = df
        self.outputs_dir = outputs_dir
        self.sample_max_rows = sample_max_rows
        os.makedirs(self.outputs_dir, exist_ok=True)

        self.protected_ids = protected_ids or []

        # Numeric & categorical (IDs excluded)
        self.numeric_cols = [
            c for c in df.select_dtypes(include=np.number).columns
            if c not in self.protected_ids
        ]
        self.categorical_cols = [
            c for c in df.select_dtypes(include=["object", "category"]).columns
            if c not in self.protected_ids
        ]

        # Global plotting style (publication-ready)
        sns.set_theme(
            style="whitegrid",
            context="notebook",
            rc={
                "axes.titlesize": 14,
                "axes.labelsize": 12,
                "xtick.labelsize": 10,
                "ytick.labelsize": 10,
                "figure.figsize": (9, 5)
            }
        )

    # --------------------------------------------------
    # VALIDATION
    # --------------------------------------------------
    def validate_columns(self, cols: List[str]):
        """
          Validate that columns exist and are not protected IDs.

          This method is a hard guardrail to prevent:
          - plotting non-existent columns
          - accidental use of identifier columns in visualizations
          """
        missing = [c for c in cols if c not in self.df.columns]
        if missing:
            raise ValueError(f"Missing required columns: {missing}")
        for c in cols:
            if c in self.protected_ids:
                raise ValueError(f"Protected ID column used in plot: {c}")

    # --------------------------------------------------
    # SAMPLING (for heavy plots)
    # --------------------------------------------------
    def sample_df_for_plot(self, df=None):
        """
        Return a safely sampled DataFrame for plotting.

        Sampling is deterministic and only applied when the dataset
        exceeds the configured row threshold.
        """
        df = df if df is not None else self.df
        if len(df) <= self.sample_max_rows:
            return df.copy()
        return df.sample(self.sample_max_rows, random_state=42).copy()

    # --------------------------------------------------
    # AXIS FORMATTING (CRITICAL)
    # --------------------------------------------------
    def _format_axis(
        self,
        ax,
        xlabel=None,
        ylabel=None,
        title=None,
        rotate_xticks=False,
        is_percent=False,
        is_currency=False,
        log_scale=False
    ):
        """
        Apply consistent axis formatting rules to Matplotlib axes.

        Centralizes formatting logic so that all plots adhere to the
        same visual and semantic standards.
        """
        if title:
            ax.set_title(title, pad=10)
        if xlabel:
            ax.set_xlabel(xlabel)
        if ylabel:
            ax.set_ylabel(ylabel)

        if rotate_xticks:
            plt.setp(ax.get_xticklabels(), rotation=30, ha="right")

        if is_percent:
            ax.yaxis.set_major_formatter(
                plt.FuncFormatter(lambda y, _: f"{y:.0f}%")
            )

        if is_currency:
            ax.xaxis.set_major_formatter(
                plt.FuncFormatter(lambda x, _: f"₹{x/1e5:.1f}L")
            )

        if log_scale:
            ax.set_xscale("log")

    # --------------------------------------------------
    # SAFE MATPLOTLIB WRAPPER
    # --------------------------------------------------
    def safe_matplotlib(self, plot_func, save_name=None):
        """
        Execute a Matplotlib plotting function safely.

        Ensures:
        - clean figure lifecycle
        - layout tightening
        - optional file persistence
        - graceful failure without breaking the notebook
        """
        try:
            plt.clf()
            plt.close("all")

            fig, ax = plt.subplots()
            plot_func(ax)

            plt.tight_layout()

            if save_name:
                path = os.path.join(
                    self.outputs_dir,
                    save_name if save_name.endswith(".png") else f"{save_name}.png"
                )
                plt.savefig(path, dpi=160)
                print(f"Saved plot: {path}")

            plt.show()

        except Exception as e:
            print("Plot failed:", e)

    # --------------------------------------------------
    # SAFE PLOTLY WRAPPER
    # --------------------------------------------------
    def safe_plotly(self, fig_func):
        """
        Execute a Plotly figure generator safely.

        Prevents runtime failures in interactive visualizations
        from stopping notebook execution.
        """
        try:
            fig = fig_func()
            if fig is not None:
                fig.show()
        except Exception as e:
            print("Plotly failed:", e)

    def _format_percent_axis(self, ax):
      """Format x-axis as percentage."""
      ax.xaxis.set_major_formatter(
          FuncFormatter(lambda x, _: f"{int(x)}%")
      )

    def _add_dti_thresholds(self, ax):
          """Add standard DTI risk thresholds.
            Overlay standard Debt-to-Income risk thresholds on plots."""
          thresholds = {
              30: "Safe",
              40: "Caution",
              50: "High Risk"
          }
          for val, label in thresholds.items():
              ax.axvline(
                  val,
                  linestyle="--",
                  linewidth=1,
                  alpha=0.7
              )
              ax.text(
                  val + 0.5,
                  ax.get_ylim()[1] * 0.9,
                  label,
                  rotation=90,
                  fontsize=9,
                  alpha=0.8
              )
    def _beautify_payment_behaviour(self, s: pd.Series) -> pd.Series:
        """
        Convert encoded Payment_Behaviour values into human-readable labels.
        """
        mapping = {
            "high_spent_large_value_payments": "High Spend – Large Value",
            "high_spent_medium_value_payments": "High Spend – Medium Value",
            "high_spent_small_value_payments": "High Spend – Small Value",
            "low_spent_large_value_payments": "Low Spend – Large Value",
            "low_spent_medium_value_payments": "Low Spend – Medium Value",
            "low_spent_small_value_payments": "Low Spend – Small Value",
        }

        return s.map(mapping).fillna(s)
# ================================================================
# Class 8: Univariate EDA
# ================================================================

class UnivariateEDA(BaseEDA):
    """
    Risk-driven Univariate EDA.
    9 charts.
    Risk-driven Univariate Exploratory Data Analysis.

    This class is responsible for analyzing the distributional
    characteristics of individual variables in isolation.

    Design intent:
    - Focus on single-variable risk signals
    - Understand spread, skewness, and outliers
    - Establish baseline behavior before relational analysis
    - Produce consistent, publication-ready plots

    Scope:
    - Supports numeric and categorical variables
    - Generates multiple plot types per variable
    - Used to build foundational intuition for later Bivariate
      and Multivariate analysis

    """

    def plot(
        self,
        column: str,
        plot_type: Union[str, List[str]],
        bins: int = 30,
        clip_upper: Optional[float] = None,
        order: Optional[List[str]] = None
    ):
        """
        Generate one or more univariate plots for a given column.

        Parameters
        ----------
        column : str
            Name of the column to be analyzed.
        plot_type : str or list of str
            Type(s) of plot to generate.
            Supported values include:
            - count
            - hist
            - kde
            - hist_log
            - box
        bins : int
            Number of bins for histogram-based plots.
        clip_upper : float, optional
            Upper bound for clipping extreme values to improve
            visualization readability.
        order : list of str, optional
            Explicit category ordering for count plots.
        """
         # Validate column existence and ID safety
        self.validate_columns([column])

         # Sample data defensively for plotting
        s = self.sample_df_for_plot()[column].copy()

        # Numeric coercion (safe, non-destructive)
        is_numeric = pd.api.types.is_numeric_dtype(s)
        if is_numeric:
            s = pd.to_numeric(s, errors="coerce")

        # Remove missing values for clean visualization
        s = s.dropna()

        # Optional clipping to control extreme tails
        if clip_upper is not None:
            s = s.clip(upper=clip_upper)


        # Normalize plot_type input to list
        plot_types = [plot_type] if isinstance(plot_type, str) else plot_type

        # ---------------------------
        # Plot executor
        # ---------------------------
        def _plot(ax):
            """
            Internal plotting routine executed inside safe_matplotlib.
            """
            for p in plot_types:

                if p == "count":

                  # Apply risk palette ONLY for Credit_Score
                  if column == "Credit_Score":
                      sns.countplot(
                          x=s.astype(str),
                          order=order,
                          palette=self.RISK_PALETTE,
                          ax=ax
                      )
                  else:
                      # Neutral color for non-risk discrete variables
                      sns.countplot(
                          x=s.astype(str),
                          order=order,
                          color=self.NUMERIC_COLOR,
                          ax=ax
                      )

                  self._format_axis(
                      ax,
                      xlabel=column,
                      ylabel="Count",
                      title=f"Distribution of {column}",
                      rotate_xticks=True
                  )

                elif p == "hist":
                    sns.histplot(s, bins=bins, stat="density", color=self.NUMERIC_COLOR, alpha=0.6, ax=ax)
                    self._format_axis(
                        ax,
                        xlabel=column,
                        ylabel="Frequency",
                        title=f"Distribution of {column}",
                        is_currency=("Income" in column)
                    )

                elif p == "kde":
                    sns.kdeplot(s, fill=False, linewidth=2.2, bw_adjust=0.6, color=self.NUMERIC_KDE_COLOR, zorder=5, ax=ax)
                    self._format_axis(
                        ax,
                        xlabel=column,
                        ylabel="Density",
                        title=f"Density of {column}"
                    )
                    # Prevent scientific notation for density axis (readability for exec review)
                    ax.ticklabel_format(style="plain", axis="y")

                elif p == "hist_log":
                    sns.histplot(s, bins=bins, ax=ax)
                    self._format_axis(
                        ax,
                        xlabel=f"log({column})",
                        ylabel="Frequency",
                        title=f"Log Distribution of {column}",
                        log_scale=True
                    )

                elif p == "box":
                    sns.boxplot(x=s, ax=ax, color=self.NUMERIC_COLOR)
                    self._format_axis(
                        ax,
                        xlabel=column,
                        title=f"Boxplot of {column}"
                    )

                else:
                    raise ValueError(f"Unsupported plot type: {p}")

            #  DOMAIN-AWARE POST-PROCESSING (ONCE PER AXIS)
            if column == "Debt_To_Income_Ratio":
                self._format_percent_axis(ax)
                self._add_dti_thresholds(ax)

        # Persist plot with deterministic naming
        save_name = f"univariate_{column}_{'_'.join(plot_types)}"
        self.safe_matplotlib(_plot, save_name)
# ================================================================
# Class 9: Bivariate EDA
# ================================================================
class BivariateEDA(BaseEDA):
    """
    Risk-driven Bivariate Exploratory Data Analysis.

    This class analyzes pairwise relationships between variables with
    explicit emphasis on credit risk interpretation.

    Design principles:
    - Credit_Score is treated strictly as an ordinal risk variable
    - Focus on how one variable conditions or shifts another
    - Emphasize distributional separation and risk amplification
    - Avoid leakage by enforcing identifier protection

    Scope:
    - Numeric vs categorical
    - Numeric vs numeric
    - Risk zone visualization via heatmaps

    7 chatrts
    """

    def plot(
        self,
        x: str,
        y: str,
        plot_type: Union[str, List[str]] = "box",
        bins_x: Optional[List[float]] = None,
        bins_y: Optional[List[float]] = None,
        target_col: str = "Has_Default"
    ):
        """
        Generate bivariate plots for two variables.

        Parameters
        ----------
        x : str
            Feature plotted on the x-axis.
        y : str
            Feature plotted on the y-axis.
        plot_type : str or list of str
            Type(s) of bivariate plot to generate.
            Supported values include:
            - box
            - violin
            - scatter
            - risk_heatmap
        bins_x : list of float, optional
            Bin edges for x-axis (required for risk_heatmap).
        bins_y : list of float, optional
            Bin edges for y-axis (required for risk_heatmap).
        target_col : str
            Binary target used for risk aggregation in heatmaps.
        """
        # --------------------------------------------------
        # Validation
        # --------------------------------------------------
        # Validate column existence and ID safety
        self.validate_columns([x, y])

        if x in self.protected_ids or y in self.protected_ids:
            raise ValueError("Protected ID column used in plot.")


        # risk_heatmap requires an explicit target column
        if plot_type == "risk_heatmap":
            self.validate_columns([target_col])

        # --------------------------------------------------
        # Prepare dataframe safely
        # --------------------------------------------------
        # Sample defensively and drop missing values
        df = self.sample_df_for_plot()[[x, y] + ([target_col] if plot_type == "risk_heatmap" else [])].copy()
        df = df.dropna()

        # --------------------------------------------------
        # Beautify categorical labels
        # --------------------------------------------------
        # Improve readability for Payment_Behaviour categories
        if x == "Payment_Behaviour":
            df[x] = self._beautify_payment_behaviour(df[x])

        # --------------------------------------------------
        # Enforce ordinal Credit_Score
        # --------------------------------------------------
        # Credit_Score is treated as an ordered risk variable
        if y == "Credit_Score":
            df[y] = pd.Categorical(
                df[y],
                categories=["poor", "standard", "good"],
                ordered=True
            )

        # Normalize plot_type input to list
        plot_types = [plot_type] if isinstance(plot_type, str) else plot_type

        # --------------------------------------------------
        # Plot executor
        # --------------------------------------------------
        def _plot(ax):
            """
            Internal plotting routine executed inside safe_matplotlib.
            """
            for p in plot_types:

                # ---------------------------
                # BOXPLOT
                # ---------------------------
                if p == "box":
                    sns.boxplot(
                        data=df,
                        x=x,
                        y=y,
                        showfliers=True,
                        ax=ax
                    )
                    self._format_axis(
                        ax,
                        xlabel=x.replace("_", " "),
                        ylabel=y.replace("_", " "),
                        title=f"{y.replace('_', ' ')} vs {x.replace('_', ' ')}",
                        rotate_xticks=True
                    )

                # ---------------------------
                # VIOLIN
                # ---------------------------
                elif p == "violin":
                    sns.violinplot(
                        data=df,
                        x=x,
                        y=y,
                        inner="quartile",
                        cut=0,
                        ax=ax
                    )
                    self._format_axis(
                        ax,
                        xlabel=x.replace("_", " "),
                        ylabel=y.replace("_", " "),
                        title=f"{y.replace('_', ' ')} vs {x.replace('_', ' ')}",
                        rotate_xticks=True
                    )

                # ---------------------------
                # SCATTER (sampled)
                # ---------------------------
                elif p == "scatter":
                    sample_df = df.sample(min(3000, len(df)), random_state=42)
                    sns.scatterplot(
                        data=sample_df,
                        x=x,
                        y=y,
                        alpha=0.4,
                        ax=ax
                    )
                    self._format_axis(
                        ax,
                        xlabel=x.replace("_", " "),
                        ylabel=y.replace("_", " "),
                        title=f"{y.replace('_', ' ')} vs {x.replace('_', ' ')} (Sampled)"
                    )

                # ---------------------------
                # RISK ZONE HEATMAP
                # ---------------------------
                elif p == "risk_heatmap":
                    if bins_x is None or bins_y is None:
                        raise ValueError("bins_x and bins_y must be provided for risk_heatmap")

                    # Discretize axes into risk bins
                    df["_x_bin"] = pd.cut(df[x], bins=bins_x, include_lowest=True)
                    df["_y_bin"] = pd.cut(df[y], bins=bins_y, include_lowest=True)

                    # Aggregate default rate within each risk zone
                    heatmap_df = (
                        df
                        .groupby(["_y_bin", "_x_bin"])[target_col]
                        .mean()
                        .mul(100)  # %
                        .unstack()
                    )

                    sns.heatmap(
                        heatmap_df,
                        annot=True,
                        fmt=".1f",
                        cmap="Reds",
                        linewidths=0.5,
                        cbar_kws={"label": "Default Rate (%)"},
                        ax=ax
                    )

                    self._format_axis(
                        ax,
                        xlabel=x.replace("_", " "),
                        ylabel=y.replace("_", " "),
                        title="Credit Risk Zones: Income vs Outstanding Debt"
                    )

                else:
                    raise ValueError(f"Unsupported plot type: {p}")

        # --------------------------------------------------
        # Safe render
        # --------------------------------------------------
        save_name = f"bivariate_{x}_vs_{y}_{'_'.join(plot_types)}"
        self.safe_matplotlib(_plot, save_name)
# ================================================================
# Class 10: Multivariate EDA
# ================================================================

class MultivariateEDA(BaseEDA):
    """
    Risk-driven Multivariate Exploratory Data Analysis.

    This class is used strictly for confirmation, not discovery.

    Design intent:
    - Validate and reinforce patterns identified in Univariate and Bivariate EDA
    - Examine interaction effects and structural stability
    - Avoid overfitting narratives by limiting multivariate plots
    - Communicate complex risk dynamics to senior stakeholders

    Scope:
    - Feature interaction diagnostics
    - Risk compression and density analysis
    - Static plots for auditability
    - Selective interactive overlays for executive exploration

    6 charts
    """

    def plot(
        self,
        cols,
        plot_type,
        x=None,
        y=None,
        hue=None,
        size=None,
        interactive=False
    ):
        """
        Generate multivariate plots to confirm risk structure.

        Parameters
        ----------
        cols : list
            Columns required for the selected plot(s).
        plot_type : str or list of str
            Type(s) of multivariate plot to generate.
            Supported values include:
            - pair
            - facet_risk
            - corr
            - parallel
            - kde
            - joint_kde
        x : str, optional
            Feature mapped to x-axis where applicable.
        y : str, optional
            Feature mapped to y-axis where applicable.
        hue : str, optional
            Risk segmentation variable (e.g., Credit_Score encoded).
        size : str, optional
            Reserved for future extensions.
        interactive : bool
            Whether to render selected interactive Plotly overlays.
        """
        # --------------------------------------------------
        # Validation
        # --------------------------------------------------
        # Validate column existence and enforce ID protection
        self.validate_columns(cols)

        for c in cols:
            if c in self.protected_ids:
                raise ValueError(f"Protected ID used: {c}")

        # Sample defensively and remove missing values
        df = self.sample_df_for_plot()[cols].dropna()
        plot_types = [plot_type] if isinstance(plot_type, str) else plot_type

        # ==================================================
        # PAIRPLOT — FIGURE LEVEL
        # ==================================================
        # Used for interaction sanity checks and model stability intuition
        if "pair" in plot_types:
            sns.pairplot(
                df,
                corner=True,
                diag_kind="kde",
                plot_kws={"alpha": 0.25, "s": 12}
            )
            plt.suptitle(
                "Feature Interaction & Model Stability Check",
                y=1.02,
                fontsize=14,
                fontweight="bold"
            )
            plt.show()
            return

        # ==================================================
        # FACET — RISK COMPRESSION
        # ==================================================
        # Shows how relationships compress or stabilize across risk segments
        if "facet_risk" in plot_types:
            g = sns.FacetGrid(
                df,
                col=hue,
                col_wrap=3,
                height=3.2,
                sharex=True,
                sharey=True
            )

            g.map_dataframe(
                sns.scatterplot,
                x=x,
                y=y,
                alpha=0.15,
                s=12,
                color="#4C72B0"
            )

            for key, ax in g.axes_dict.items():
                df_seg = df[df[hue] == key]

                sns.regplot(
                    data=df_seg,
                    x=x,
                    y=y,
                    scatter=False,
                    lowess=True,
                    color="black",
                    line_kws={"linewidth": 2},
                    ax=ax
                )

                # Debt stress reference thresholds
                ax.axhline(1000, linestyle="--", color="gray", alpha=0.6)
                ax.axhline(3000, linestyle="--", color="gray", alpha=0.6)

            g.fig.suptitle(
                "Debt–Income Stress Compresses as Credit Quality Improves",
                y=1.08,
                fontsize=14,
                fontweight="bold"
            )

            plt.show()
            return

        # ==================================================
        # AXIS-BASED PLOTS
        # ==================================================
        def _plot(ax):
            """
            Internal plotting routine executed inside safe_matplotlib.
            """
            for p in plot_types:

                # -------------------------
                # Correlation Heatmap
                # -------------------------
                if p == "corr":
                    sns.heatmap(
                        df.corr(),
                        cmap="RdBu",
                        center=0,
                        annot=False,
                        ax=ax
                    )
                    ax.set_title("Correlation Heatmap (Top Risk Drivers)", fontweight="bold")

                # -------------------------
                # Parallel — MEDIAN PROFILES (STATIC)
                # -------------------------
                elif p == "parallel":
                    df_pc = df.copy()

                    # Map encoded risk levels to readable segments
                    df_pc["Risk_Segment"] = df_pc[hue].map({
                        1: "Poor",
                        2: "Standard",
                        3: "Good"
                    })

                    drivers = ["Annual_Income", "Outstanding_Debt", "Interest_Rate"]

                    # Normalize for cross-feature comparison
                    scaler = MinMaxScaler()
                    df_pc[drivers] = scaler.fit_transform(df_pc[drivers])

                    # Median profiles per risk segment
                    df_median = (
                        df_pc.groupby("Risk_Segment")[drivers]
                        .median()
                        .reset_index()
                    )

                    parallel_coordinates(
                        df_median,
                        class_column="Risk_Segment",
                        cols=drivers,
                        linewidth=4,
                        ax=ax,
                        color=["#C44E52", "#4C72B0", "#DD8452"]
                    )

                    ax.set_title(
                        "Median Risk Profiles Separate Clearly by Credit Quality",
                        fontsize=13,
                        fontweight="bold"
                    )

                    ax.set_ylabel("Normalized Median Value (0–1)")
                    ax.set_ylim(0, 1)
                    ax.grid(axis="y", alpha=0.3)

                # -------------------------
                # KDE — SCORE-CONDITIONED DISTRIBUTION
                # -------------------------
                elif p == "kde":
                    sns.kdeplot(
                        data=df,
                        x="Outstanding_Debt",
                        hue=hue,
                        fill=True,
                        common_norm=False,
                        alpha=0.45,
                        linewidth=2,
                        ax=ax
                    )

                    ax.set_title(
                        "Debt Distribution Tightens as Credit Quality Improves",
                        fontsize=13,
                        fontweight="bold"
                    )

                    ax.set_xlabel("Outstanding Debt")
                    ax.set_ylabel("Density")
                # -------------------------
                # Joint KDE — Risk Density Zones
                # -------------------------
                elif p == "joint_kde":

                    if hue is None:
                        raise ValueError("joint_kde requires `hue` (credit score)")

                    # Color palette (same as Chart 21 & 22)
                    palette = {
                        1: "#C44E52",   # Poor
                        2: "#DD8452",   # Standard
                        3: "#4C72B0"    # Good
                    }

                    # Draw KDE contours per credit segment
                    for score, color in palette.items():
                        df_seg = df[df[hue] == score]

                        sns.kdeplot(
                            data=df_seg,
                            x=x,
                            y=y,
                            fill=True,
                            alpha=0.35,
                            levels=6,
                            thresh=0.05,
                            color=color,
                            ax=ax
                        )

                    # Executive framing
                    ax.set_title(
                        "Risk Density Concentrates and Compresses as Credit Quality Improves",
                        fontsize=13,
                        fontweight="bold"
                    )

                    ax.set_xlabel(x.replace("_", " "))
                    ax.set_ylabel(y.replace("_", " "))

                    # Debt stress reference bands (same language as Chart 20)
                    ax.axhline(1000, linestyle="--", color="gray", alpha=0.5)
                    ax.axhline(3000, linestyle="--", color="gray", alpha=0.5)

                    ax.text(0.02, 0.82, "High Debt Stress",
                            transform=ax.transAxes, fontsize=9, color="gray")
                    ax.text(0.02, 0.55, "Moderate Debt",
                            transform=ax.transAxes, fontsize=9, color="gray")
                    ax.text(0.02, 0.25, "Low Debt",
                            transform=ax.transAxes, fontsize=9, color="gray")

                else:
                    raise ValueError(f"Unsupported plot type: {p}")

        # ---- STATIC RENDER
        save_name = f"multivariate_{'_'.join(plot_types)}"
        self.safe_matplotlib(_plot, save_name)


        # ==================================================
        # ==================================================
        # INTERACTIVE OVERLAYS
        # ==================================================
        if interactive:

            # ---- Interactive Parallel (CORRECT & STABLE)
            if "parallel" in plot_types:
                fig = px.parallel_coordinates(
                    df,
                    dimensions=[
                        "Annual_Income",
                        "Outstanding_Debt",
                        "Interest_Rate"
                    ],
                    color=hue,
                    range_color=[
                        df[hue].min(),
                        df[hue].max()
                    ],
                    color_continuous_scale=px.colors.sequential.Blues,
                    labels={
                        "Annual_Income": "Annual Income",
                        "Outstanding_Debt": "Outstanding Debt",
                        "Interest_Rate": "Interest Rate (%)",
                        hue: "Credit Score Band"
                    },
                    title="Parallel Coordinates — Multivariate Risk Profile Validation"
                )

                # IMPORTANT: No hovertemplate for parcoords
                fig.update_layout(
                    coloraxis_colorbar=dict(
                        title="Credit Score Band"
                    )
                )

                fig.show()

            # ---- Interactive KDE (unchanged, already correct)
            if "kde" in plot_types:
                fig = px.histogram(
                    df,
                    x="Outstanding_Debt",
                    color=hue,
                    nbins=60,
                    opacity=0.6,
                    marginal="rug",
                    labels={
                        "Outstanding_Debt": "Outstanding Debt",
                        hue: "Credit Score Band"
                    },
                    title="Interactive Debt Distribution by Credit Quality"
                )
                fig.show()


print("EDA module loaded: UnivariateEDA, BivariateEDA, MultivariateEDA, FeatureEngineering")

````
